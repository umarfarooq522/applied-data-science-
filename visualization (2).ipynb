```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("adult.csv")
df.head()
```

```python
def make_lineplot(feature1, feature2, title):
    plt.figure(figsize=(12, 5))
    plt.plot(df.index[:100], df[feature1][:100], label=feature1)
    plt.plot(df.index[:100], df[feature2][:100], label=feature2)
    plt.title(title)
    plt.legend()
    plt.show()

make_lineplot("hours-per-week", "age", title="Peple's Age VS Work Hour")
```

```python
def make_pieplot(data, label, title):
    plt.figure(figsize=(15, 5))
    plt.pie(x=data, labels=label, explode=[0, 0.1])
    plt.title(title)
    plt.show()

    
temp = df.gender.value_counts()
make_pieplot(temp.values, temp.index, "Distribution of Gender")
```

```python
def make_barplot(x, y, title):
    plt.figure(figsize=(15, 5))
    plt.bar(x=x, height=y)
    plt.title(title)
    plt.xlabel("Education")
    plt.ylabel("Hours per Week")
    plt.xticks(rotation=90)
    plt.ylim(25, 50)
    plt.show()

temp = df.groupby("education")["hours-per-week"].mean().sort_values()
make_barplot(temp.index, temp.values, "Average Work Hour by Education Level")
```
